```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data1.csv')
df
```

```python
fig, ax1 = plt.subplots(figsize=(5, 3))
ax1.bar(df['epoch'], df['detector_confidence'], color='forestgreen', label='Detector Confidence')
ax1.set_xlabel('Model on Epoch')
ax1.set_ylabel('Detector Confidence', color='forestgreen')
ax1.tick_params(axis='y', labelcolor='forestgreen')

ax2 = ax1.twinx()
ax2.plot(df['epoch'], df['si_sdr'], color='red', label='Average SI-SDR', marker='o')
ax2.set_ylabel('Average SI-SDR', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.ticklabel_format(axis='y', style='plain')

ax3 = ax1.twinx()
ax3.spines["right"].set_position(("outward", 40))  # shift outward
ax3.plot(df['epoch'], df['avg_pesq'],
         color='royalblue', label='Average PESQ', marker='o')
ax3.set_ylabel('Average PESQ', color='royalblue')
ax3.tick_params(axis='y', labelcolor='royalblue')

plt.title('Model Performance Over Epochs')
# fig.legend(bbox_to_anchor=(0.92, 0.7725), fontsize=9, framealpha=0, handlelength=0.75)
plt.savefig('model_performance.png', bbox_inches='tight')
plt.show()
```
